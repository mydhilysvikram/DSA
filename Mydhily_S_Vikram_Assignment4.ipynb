```python
import pandas as pd
import numpy as np
```

1. Load the dataset into python environment

```python
titanic_data = pd.read_csv(r'C:\Users\user\DSA\Assignments\Assignment 4\titanic_dataset.csv')
titanic_data
```

2. Make ‘PassengerId’ as the index column

```python
titanic_data.set_index('PassengerId', inplace=True)
titanic_data
```

3. Check the basic details of the dataset

```python
titanic_data.info()

print(titanic_data.describe())

print(titanic_data.shape)
```

4. Fill in all the missing values present in all the columns in the dataset

```python
# from titanic_data.info() it is clear that missing values are present in the columns 'Age','Cabin' and 'Embarked'
# therefore filling those columns only

titanic_data['Embarked'] = titanic_data['Embarked'].fillna('Q')
titanic_data['Cabin'] = titanic_data['Cabin'].fillna('C23 C25 C27')
titanic_data['Age'].fillna(titanic_data['Age'].median(), inplace=True)

# checking if there is any missing values again

titanic_data.info()
```

5. Check and handle outliers in at least 3 columns in the dataset

```python
import matplotlib.pyplot as plt
```

```python
# checking outliers using box plot

plt.boxplot(titanic_data['Survived'], vert=True)
plt.xlabel('Survived')
plt.title('Box Plot of Survived')
plt.show()

plt.boxplot(titanic_data['SibSp'], vert=True)
plt.xlabel('SibSp')
plt.title('Box Plot of SibSp')
plt.show()

plt.boxplot(titanic_data['Fare'], vert=True)
plt.xlabel('Fare')
plt.title('Box Plot of Fare')
plt.show()

```

 'Survived' column has no outliers

```python
# handling outliers in 'SibSp' column using Z score

upper_limit = titanic_data['SibSp'].mean()+ 3* titanic_data['SibSp'].std()
lower_limit = titanic_data['SibSp'].mean()- 3* titanic_data['SibSp'].std()

print(upper_limit)
print(lower_limit)

titanic_data.loc[titanic_data['SibSp'] >= upper_limit, 'SibSp'] = 2.5
```

```python
# handling outliers in 'Fare' column using Z score

upper_limit = titanic_data['Fare'].mean()+ 3* titanic_data['Fare'].std()
lower_limit = titanic_data['Fare'].mean()- 3* titanic_data['Fare'].std()

print(upper_limit)
print(lower_limit)

titanic_data.loc[titanic_data['Fare'] >= upper_limit, 'Fare'] = 60
```

6. Do min max scaling on the feature set (Take ‘Survived’ as target)

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
titanic_data[['Survived']] = scaler.fit_transform(titanic_data[['Survived']])
```
