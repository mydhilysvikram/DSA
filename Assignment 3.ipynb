# 2. Case Study on Testing of Hypothesis

```python
from scipy.stats import ttest_ind
from scipy.stats import chi2_contingency
```

```python
sales_data  = pd.read_csv(r'C:\Users\user\DSA\Datasets\Sales_add.csv')
sales_data
```

# a. The company wishes to clarify whether there is any increase in sales after stepping into digital marketing.

```python
t,pv = ttest_ind(sales_data['Sales_before_digital_add(in $)'],sales_data['Sales_After_digital_add(in $)'])
```

```python
if pv<0.05:
    print('Reject Null Hypothesis')
else:
    print('Accept Null Hypothesis')
```

Observation : There is an increase in sales after the company moved into digital marketing

# b. The company needs to check whether there is any dependency between the features â€œRegionâ€ and â€œManagerâ€.

```python
new_data = pd.crosstab(sales_data['Region'],sales_data['Manager'])
```

```python
chi, p_v,_,_ = chi2_contingency(new_data)
```

```python
if p_v<0.05:
    print("Rejecting Independence")
else:
    print("Accepting Independence")
```

Observation : It shows us that "Region" and "Manager" are independent of each other.
